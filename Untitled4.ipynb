```python
df['popularity'] = df['streams'] / df['rank']
```

```python
pivot_table = df.pivot_table(index='title', columns='artist', values='streams', aggfunc='sum', fill_value=0)
```

```python
print(pivot_table) 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
pivot_table_scaled = scaler.fit_transform(pivot_table)
print(pivot_table_scaled) 
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=50, random_state=42)
latent_matrix = svd.fit_transform(pivot_table_scaled)
```

```python
print(latent_matrix)  
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(latent_matrix)
```

```python
print(similarity)  
```

```python
song_indices = pd.Series(pivot_table.index, index=pivot_table.index).to_dict()
```

```python
def recommend_songs(song_title, similarity_matrix, pivot_table):
    # Step 1: Find the index of the song in pivot_table
    song_index = pivot_table.index.get_loc(song_title)
    
    # Step 2: Get similarity scores for the song
    similarity_scores = similarity_matrix[song_index]
    
    # Step 3: Pair scores with song indices and sort by similarity score
    similar_songs = list(enumerate(similarity_scores))
    similar_songs_sorted = sorted(similar_songs, key=lambda x: x[1], reverse=True)
    
    # Step 4: Extract indices of recommended songs
    recommended_song_indices = [x[0] for x in similar_songs_sorted]
    
    # Step 5: Get titles of recommended songs (excluding the input song itself)
    recommended_songs = [pivot_table.index[i] for i in recommended_song_indices if i != song_index]
    
    return recommended_songs

# Example usage:
song_title = "Yellow"
recommended_songs = recommend_songs(song_title, similarity, pivot_table)
print(f"Recommended songs for '{song_title}':")
print(recommended_songs[:10])  # Print top 10 recommended songs
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Recommended songs list
recommended_songs = ['Christmas Lights', 'The Scientist', 'Higher Power', 'Vivo Pensando En Ti (feat. Maluma)', 'TEEN KASIA', 'PLEASE', 'ปล่อยปาก', 'Best Friend (feat. NERVO, The Knocks & ALISA UENO)', 'Jólastjarnan', 'Токсичен']

# Filter data for recommended songs
recommended_df = df[df['title'].isin(recommended_songs)]

# Extract relevant columns for analysis
recommended_data = recommended_df[['title', 'streams', 'rank', 'popularity']].set_index('title')

print(recommended_data) 

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=recommended_data.index, y='streams', data=recommended_data, palette='viridis')
plt.title('Streams Comparison for Recommended Songs')
plt.xlabel('Songs')
plt.ylabel('Streams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
sns.barplot(x=recommended_data.index, y='rank', data=recommended_data, palette='plasma')
plt.title('Rank Comparison for Recommended Songs')
plt.xlabel('Songs')
plt.ylabel('Rank')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
sns.barplot(x=recommended_data.index, y='popularity', data=recommended_data, palette='magma')
plt.title('Popularity Comparison for Recommended Songs')
plt.xlabel('Songs')
plt.ylabel('Popularity (Streams per Rank)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
